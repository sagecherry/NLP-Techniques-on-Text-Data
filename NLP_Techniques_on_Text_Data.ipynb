```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
text='Natural language processing is interesting'
tokens=word_tokenize(text)
print(tokens)
```

```python
#Sentence Tokenization
from nltk.tokenize import sent_tokenize
text="Python is easy. It is widely used in data science"
sentences=sent_tokenize(text)
print(sentences)
```

```python
#Stop Word Removal
#stop words such as is/the/and/in are removed
from nltk.corpus import stopwords
stop_words=set(stopwords.words('english'))
words=word_tokenize(text)
filtered_words=[word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

```python
#Stemming
#it reduces words to their room form
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
words=['playing','runs','swimming']
for w in words:
  print(stemmer.stem(w))
```

```python
#Lemmatization
#Lemmatization converts words to their dictionary form.
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
print(lemmatizer.lemmatize('buddies'))
print(lemmatizer.lemmatize('studies'))
```

```python
#Part of Speech (POS)Tagging
#POS tagging indentifies grammatical roles of words
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
words=word_tokenize('Python is a powerfull programming language')
pos_tags=pos_tag(words)
print(pos_tags)
```

```python
#Word Frequency Count
from nltk.probability import FreqDist
words=word_tokenize(text)
freq_dist=FreqDist(words)
print(freq_dist.most_common())
```
